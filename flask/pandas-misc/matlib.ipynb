```python
# https://www.youtube.com/watch?v=JpSMse3eVVg
# Creating Visualizations using Pandas Library | Python Pandas Tutorials

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


```

```python
df = pd.read_csv("https://raw.githubusercontent.com/AlexTheAnalyst/PandasYouTubeSeries/refs/heads/main/Ice%20Cream%20Ratings.csv")
df = df.set_index("Date")
df
```

```python
df.plot(kind = 'line')
```

```python
df['Flavor Rating'].plot(kind = 'bar', stacked=True)
```

```python
df.plot(kind = 'barh', stacked=True)
```

```python
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating')
```
